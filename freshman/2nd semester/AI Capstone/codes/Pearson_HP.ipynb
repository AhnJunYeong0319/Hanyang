```python
from xgboost import XGBRegressor, plot_importance 
import pandas as pd
import numpy as np
import math
from sklearn.model_selection import cross_validate, train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn import metrics
import matplotlib.pyplot as plt
import time                                                                                              
data = pd.read_csv('Pearson_data.csv')
data = data.drop(data.columns[3], axis=1)
data = data.drop(data.columns[3], axis=1)
data = data.drop(data.columns[7], axis=1)
data = data.fillna(data.mean())
#display(data.head())

x = data[['sim_user2','sim_user3','sim_user4','sim_user5', 'sim_movie2','sim_movie3','sim_movie4','sim_movie5']]
col_names = ['sim_user2','sim_user3','sim_user4','sim_user5', 'sim_movie2','sim_movie3','sim_movie4','sim_movie5']
y = data['rating']

scorers = { 'RMSE': 'neg_mean_squared_error', 'MAE': 'neg_mean_absolute_error'}

Eta = [0.1, 0.3, 0.5, 0.7]
Gamma = [0, 1, 3, 5]
Max_depth = [4, 6, 8, 10]
Min_child_weight = [1,3,5,7]

RMSEs = []
MAEs = []

for e in Eta:
    for g in Gamma:
        for d in Max_depth:
            for c in Min_child_weight:
                start_ts=time.time()
                rgr = XGBRegressor(eta=e, gamma=g, max_depth=d, min_child_weight=c)
                scores = cross_validate(rgr, x, y,scoring=scorers, cv=5)
                rmse_score = np. asarray([math.sqrt(-x) for x in scores['test_RMSE']])
                rmse_mean = rmse_score.mean() 
                rmse_std = rmse_score.std()
                mae_score = np.asarray(-scores['test_MAE'])
                mae_mean = mae_score.mean()
                mae_std = mae_score.std()

                print("="*50)
                print("Eta:{}, Gamma:{}, Max_depth:{}, Min_child_weight:{}".format(e,g,d,c))
                print("\n")
                print("XGBoost RMSE : {0:0.3f} (+/- {1:0.3f})".format(rmse_mean,rmse_std * 2))
                print("XGBoost MAE  : {0:0.3f} (+/- {1:0.3f})".format(mae_mean,mae_std * 2))
                print("CV Runtime   : {0:0.3f}".format(time.time()-start_ts)) 
                print("="*50)        
                
                RMSEs.append(rmse_mean)
                MAEs.append(mae_mean)
```

```python
Pear_scores = pd.DataFrame(MAEs, RMSEs)
Pear_scores.reset_index(inplace=True)

Eta = [0.1, 0.3, 0.5, 0.7]
Gamma = [0, 1, 3, 5]
Max_depth = [4, 6, 8, 10]
Min_child_weight = [1,3,5,7]


indices = []

for e in Eta:
    for g in Gamma:
        for d in Max_depth:
            for c in Min_child_weight:
                indices.append('{}/{}/{}/{}'.format(e,g,d,c))
                
indices = pd.Series(indices)
Pear_HP = pd.concat([indices, Pear_scores], axis=1)
Pear_HP.columns = ['Params(e/g/d/c)', 'RMSE', 'MAE']
Pear_HP
```

```python
Pear_HP.to_csv('Pear_HP.csv', index=False)
```
