```python
import pandas as pd
import numpy as np

movie_sim = pd.read_csv("out.csv")
user = pd.read_csv("ratings.csv")
movie = pd.read_csv("movies.csv")

movie_sim = movie_sim.drop([movie_sim.columns[0]], axis=1)
movie_sim['movieId'] = movie['movieId']

print(movie_sim)

movie_id = max(user['movieId'])+1
user_id = max(user['userId'])+1

movie_check =  np.zeros(user_id*movie_id, dtype='uint8').reshape(user_id,movie_id)

for i in range(len(user)):
    movie_check[user['userId'][i]][user['movieId'][i]] = user['rating'][i]

f = open("feature_movie.csv", 'w')

for i in range(5):
    f.write("sim_movie"+str(i+1)+",")
f.write("\n")

len_user = len(user)

for i in range(len_user):
    target_movie = user['movieId'][i]
    target_user = user['userId'][i]
    ch_movie=0
    
    a = np.array(movie_sim[movie_sim['movieId']==target_movie])[0]
    a = a[49::-1]
    
    for j in range(50):
        if movie_check[user['userId'][i]][a[j]] != 0:
            f.write(str(movie_check[user['userId'][i]][a[j]])+",")
            ch_movie=ch_movie+1
        if ch_movie == 5:
            break

    #if ch_movie!=5:
        #for j in el:
            #f.write(str(avg)+",")
    print(i/len_user*100,"% is done")
    f.write("\n")

f.close()
```
