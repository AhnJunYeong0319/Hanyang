```python
import pandas as pd
import numpy as np
from operator import itemgetter

sampling=1  # sampling method 
check=2     # checking method 1:cosine similarity, 2:finding high simialr movies
show=0      # 1:show genre list and similarity matrix

def cosine_sim(a, b):
    return np.dot(a, b)/(np.linalg.norm(a)*np.linalg.norm(b))

movie = pd.read_csv("./movies.csv")

add_genre = []
for i in range(len(movie)):
    add_genre.append(movie['genres'][i].split("|"))

movie['genre']=add_genre
movie=movie.drop(['genres'],axis=1)

if sampling==1:
    movie_sample=pd.DataFrame(movie.sample(n=1000, random_state=1))
    movie_sample=movie_sample.reset_index()
    movie_sample=movie_sample.drop(['index'], axis=1)
    
    movie=movie_sample

genre_list = []
for i in range(len(movie)):
    genre_list=genre_list+movie['genre'][i]
    make_set=set(genre_list)
    genre_list=list(make_set)

if show == 1:
    print("=================== Genre Lists ===================","\n",genre_list)
    print("\n")

movie_matrix = [[0 for col in range(len(genre_list))] for row in range(len(movie))]

for i in range(len(movie)):
    for j in range(len(genre_list)):
        if set([genre_list[j]])&set(movie['genre'][i]):
            movie_matrix[i][j]=1
            
movie_matrix = np.array(movie_matrix)

if check==1:
    a=5
    b=8
    print(movie['genre'][a], "\n")
    print(movie_matrix[a], "\n")
    print(movie['genre'][b], "\n")
    print(movie_matrix[b], "\n")
    print(cosine_sim(movie_matrix[a], movie_matrix[b]),"\n")


movie_similarity = [[0 for col in range(len(movie))] for row in range(len(movie))]

for i in range(len(movie)):
    for j in range(len(movie)):
        movie_similarity[i][j]=cosine_sim(movie_matrix[i], movie_matrix[j])

movie_similarity = np.array(movie_similarity)

if show == 1:
    print("================ Similarity Matrix ================", "\n")
    print(movie_similarity, "\n")

if check==2:
    ch =  5           #target movie
    top = 10          #top n high similar movie
    list_sort = []
    print("\n")
    print("============ Finding High Similar Movie ============", "\n")
    print("High similarity with <%s> is : "%movie['title'][ch], "\n")
    
    index = np.arange(0, len(movie), 1)
    for i in range(len(movie)):
        arr = [index[i], movie_similarity[ch][i]]
        list_sort.append(arr)

    list_sort.sort(key=lambda x:-x[1])    
    for i in range(top):
        print("<%s>"%movie['title'][list_sort[i][0]])
        print("cosine similarity :", list_sort[i][1], "\n")
```
