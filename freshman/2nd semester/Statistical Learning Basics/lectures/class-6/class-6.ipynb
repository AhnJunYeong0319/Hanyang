```r
intercept = c()
slope = c()
```

```r
for (i in 1:1000){
    result = coef(lm(mpg ~ horsepower, data=Auto, subset=sample(392, replace=TRUE)))
    intercept = c(intercept, result[1])
    slope = c(slope, result[2])
}
```

```r
print(paste('intercept: ', round(mean(intercept),7), round(sd(intercept),7)))
print(paste('slope: ', round(mean(slope),7), round(sd(slope),7)))
```
