```r
x <- rnorm(100)
z <- rnorm(100)
```

```r
y <- z+2*x
```

```r
rm(z)
```

```r
plot(x)
```

```r
plot(density(x))
```

```r
plot(x,y, ylab = 'random y', xlab = 'random x', xlim=c(-3,3), ylim = c(-4,4))
abline(0,2)
abline(0,2,col = 'red')
abline(0.02733,2.06213,col = 'red')

exerc1 = data.frame(cbind(y,x))

ex1.lm <- lm(y ~ x, exerc1) 
abline(ex1.lm$coefficients, col='red')
```
